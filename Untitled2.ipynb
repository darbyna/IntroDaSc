```python
#Source(s) used for assistance: 
# PRIMARY SOURCE: https://docs.python.org/3/library/xml.etree.elementtree.html
# http://2017.compciv.org/guide/topics/python-nonstandard-libraries/beautifulsoup.html
# https://stackoverflow.com/questions/1912434/how-do-i-parse-xml-in-python
# https://www.geeksforgeeks.org/xml-parsing-python/
# https://www.crummy.com/software/BeautifulSoup/bs4/doc/
from bs4 import BeautifulSoup
import lxml

```

## Parse the XML file using the "lxml" package: 

```python
#SOURCE: https://www.kite.com/python/answers/how-to-pretty-print-xml-files-in-python
power_source = lxml.etree.parse(r'xml_containing_html.xml')
super_power_source = lxml.etree.tostring(power_source, encoding="unicode", pretty_print=True)
print(super_power_source)
```

## Further parse the XML file with BeautifulSoup and utilize "html.parser":

```python
#SOURCE: https://www.crummy.com/software/BeautifulSoup/bs4/doc/
ultrx = BeautifulSoup(super_power_source, 'html.parser')
```

## Now, convert the XML data to strictly HTML:

```python
#SOURCE: https://www.crummy.com/software/BeautifulSoup/bs4/doc/
HTML_Version = ultrx.get_text()
print(HTML_Version)
```

```python
#SOURCE: https://www.crummy.com/software/BeautifulSoup/bs4/doc/
YONXU = BeautifulSoup(HTML_Version, 'html.parser')
```

## Finally, extract the urls that are embedded in the "a" tag of each HTML page. Note the page that it is from next to the url: 

```python
#SOURCE: https://www.crummy.com/software/BeautifulSoup/bs4/doc/
for url in YONXU.find_all('a'):
    print("Links Per Page:",YONXU.title.string, "|", " ", url.get('href'))
```
